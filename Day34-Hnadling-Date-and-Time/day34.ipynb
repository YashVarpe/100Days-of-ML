**Working with dates**

```python
#Converting to datetime datatype

date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

2. Extract month

```python
date['date_month'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
date['date_day_of_week'] = date['date'].dt.day_of_week
date.head()
```

```python
#Name of the day
date['date_day_name'] = date['date'].dt.day_name()
date.head()
```

```python
date.head()
```

```python
#is weekend?

date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Saturday', 'Sunday']), 1, 0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
#Extract week of the year

date['date_week'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
#Extract quarter

date['date_quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
#Extract semester

date['date_semester'] = np.where(date['date_quarter'].isin([1, 2]), 1, 2)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
#Extract time elapsed between dates

import datetime
today = datetime.datetime.today()

today   #gives todays current date and time
```

```python
today - date['date']    #time elapsed between todays date and the date in the dataset
```

```python
(today - date['date']).dt.days  #days elapsed between todays date and date in the dataset
```

```python
time['date'] = pd.to_datetime(time['date'])
time.info()
```

```python
#Extracting time, minutes and seconds from date column which has date and time of msgs

time['hour'] = time['date'].dt.hour
time['minute'] = time['date'].dt.minute
time['second'] = time['date'].dt.second
```

```python
time.head()
```

```python
#Extract time part from date column

time['time'] = time['date'].dt.time
time.head()
```

```python
#time difference

(today - time['date'])

#same can be done for time difference in hours, minutes and seconds
```

```python
#time difference in seconds

(today - time['date']) / np.timedelta64(1, 's')
```

```python
#time difference in minutes

(today - time['date']) / np.timedelta64(1, 'm')
```

```python
#time difference in hours

(today - time['date']) / np.timedelta64(1, 'h')
```
