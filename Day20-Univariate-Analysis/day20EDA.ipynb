**Exploratory Data Analysis** <br><br>

Univariate Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.sample(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**1. Categorical Data**

a. Countplot - frequency count for categorical data

```python
sns.countplot(x = 'Survived', data = df)
df['Survived'].value_counts()
```

```python
sns.countplot(x='Sex', data=df)
```

b. piechart - to display categorical data in terms of percentage

```python
df['Pclass'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
df['Sex'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

**2. Numerical data**

a. Histogram

```python
plt.hist(df['Age'])
```

b. Displot

```python
#distplot is deprecated function now

sns.displot(data=df, x='Age', kind='kde')   

#kde - kernel density estimation
#the curve traced is the prob distribution function 
```

c. Boxplot - It gives you 5 number summary

    values:
0. IQR = Interquartile Range = Q3 - Q1
1. Q1 - minimum (less than 25%) = Q1 - 1.5 * IQR
2. Q2 - 25th percentile
3. Median
4. Q3 - 75th percentile
5. Q5 - Maximum (greater than 75th percentile) = Q3 + 1.5 *IQR
6. Outliers - Anomalies in the data - data that does not follow the normal data 

```python
#boxplot

sns.boxplot(x='Age', data = df)
```
